```python
# Sentiment Analysis - Basic Version

!pip install textblob
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "Review": [
        "I love this product!",
        "This is terrible, I hate it.",
        "Not bad, could be better.",
        "Amazing experience!",
        "Worst thing ever",
        "I feel neutral about this."
    ]
}
df = pd.DataFrame(data)

# Sentiment analysis
df["Sentiment Score"] = df["Review"].apply(lambda x: TextBlob(x).sentiment.polarity)
df["Sentiment"] = df["Sentiment Score"].apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))

# Show results
print(df)

# Pie chart
df["Sentiment"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sentiment Distribution")
plt.show()
```
